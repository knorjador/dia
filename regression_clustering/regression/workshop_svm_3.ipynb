```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

dataset = pd.read_csv('./Position_Salaries.csv')
```

```python

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
    
```

```python

from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.reshape(-1, 1)).ravel()
```

```python

from sklearn.svm import SVR
```

```python

def displayKernel(kernel):
    regressor = SVR(kernel=kernel)
    regressor.fit(X, y)
    #y_pred = regressor.predict(sc_X.transform(np.array([[8.5]]))).reshape(-1, 1)
    y_pred = regressor.predict([[8.5]]).reshape(-1, 1)
    y_pred = sc_y.inverse_transform(y_pred)
    print()
    print('y_pred:' + str(y_pred[0][0]))
    print()
    X_grid = np.arange(min(X), max(X), 0.01)
    X_grid = X_grid.reshape((len(X_grid), 1))
    plt.scatter(X, y, color='blue')
    plt.plot(X_grid, regressor.predict(X_grid), color='red')
    plt.title('Truth or Bluff => SVR, kernel ' + kernel)
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    
displayKernel('linear')
displayKernel('poly')
displayKernel('rbf')
```
