```python

import pandas

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

```python

data = pandas.read_csv('./play_or_not_numeric.csv')

print(data)
```

```python

# features
X = data.iloc[:,:-1]

# target
y = data.iloc[:,-1] 

X_dum = pandas.get_dummies(X)

print(X_dum.head())

estimator = DecisionTreeClassifier()

estimator.fit(X_dum, y)

export_graphviz(estimator, out_file="cart_tree.dot", feature_names = X_dum.columns, class_names=estimator.classes_)

#install graphviz at https://www.graphviz.org/download/ 
#run command line: 
#dot -Tpng cart_tree.dot -o cart_tree.png
```
