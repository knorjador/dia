```python

import re
import requests
import pandas as pd

from bs4 import BeautifulSoup

def display(to_display, message = ''):
    print()
    if (len(message ) > 0):
        print(' > ' + str(message) + ':')
        print()
    print(to_display)
    print()
```

```python

urls = []
nb_urls = 0

with open('muzeo_urls.txt', 'r') as file:
    for url in file:
        urls.append(url.strip().replace('"', ''))
        nb_urls += 1

display(nb_urls, 'Nb urls')
display(urls[0])
```

```python

# filter urls 

filtered_urls = []
nb_filtered_urls = 0

for url in urls:
    if "papier-peint-oeuvre" in url:
        splitted = url.split('#')[0]
        if splitted not in filtered_urls:
            filtered_urls.append(splitted)
            nb_filtered_urls += 1

display(nb_filtered_urls, 'Nb filtered urls')
display(filtered_urls)
```

```python

titles = []
url_images = []

infos = []

def get_infos(url):
    page = requests.get(url)
    status_code = page.status_code
    title = 'no title'
    url_image = 'no url image'
    if (status_code == 200):
        soup = BeautifulSoup(page.content, 'html.parser')
        finded_title = re.findall(r'<title>(.*?)</title>', page.text)
        if finded_title:
            title = finded_title[0].split('– Papier peint')[0]
        meta_image = soup.find('meta', property='og:image')
        if meta_image:
            url_image = meta_image['content']

    infos.append({ 

        "url": url,
        "status_code": status_code,
        "title": title,
        "url_image": url_image

    })

for url in filtered_urls:
    get_infos(url)

df = pd.DataFrame(infos)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

display(df.shape)
display(df.head(90))
```
